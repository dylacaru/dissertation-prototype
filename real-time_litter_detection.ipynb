Using both MOG2 and KNN background subtractors

```python
import os
import numpy as np
import cv2
from ultralytics import YOLO

os.environ["KMP_DUPLICATE_LIB_OK"] = "TRUE"

cap = cv2.VideoCapture("uav_rural.mp4")
model = YOLO("models/X3-C565.pt")

mog2 = cv2.createBackgroundSubtractorMOG2()
knn = cv2.createBackgroundSubtractorKNN()

roi_points = np.array([[0, 670], [1800, 400], [0, 1000], [0, 20]])

def draw_bbox(frame, left, top, right, bottom, class_id, is_detected):
    overlay = frame.copy() # create an overlay copy outside the loop
    if class_id == 4:
        cv2.rectangle(frame, (left, top), (right, bottom), (0, 255, 0), 2)
    elif class_id != 4:
        if not is_detected:
            contours, _ = cv2.findContours(
                fgMask, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE
            )
            for cnt in contours:
                M = cv2.moments(cnt)
                if M["m00"] != 0:
                    cX = int(M["m10"] / M["m00"])
                    cY = int(M["m01"] / M["m00"])
                    cv2.circle(overlay, (cX, cY), 15, (0, 165, 255), -1) # draw circles on the overlay

            opacity = 0.5
            cv2.addWeighted(overlay, opacity, frame, 1 - opacity, 0, frame)
            cv2.putText(
                frame,
                "Littering Detected!",
                (left, top),
                cv2.FONT_HERSHEY_COMPLEX,
                0.5,
                (0, 255, 255),
                1,
                cv2.LINE_AA,
            )


def mask_binary(frame):
    # Convert to HSV color space
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)

    # Define the lower and upper bounds of the road color
    lower_color = np.array([0, 0, 0])
    upper_color = np.array([225, 60, 200])

    # Threshold the image to extract the road area
    mask = cv2.inRange(hsv, lower_color, upper_color)

    # Apply a morphological operation to remove noise and fill gaps
    kernel = np.ones((12, 12), np.uint8)
    mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)

    return mask


while True:
    ret, frame = cap.read()
    if not ret:
        break

    frame_mask = np.zeros_like(frame)

    cv2.fillPoly(frame_mask, [roi_points], (255, 255, 255))

    masked_frame = cv2.bitwise_and(frame, frame_mask)

    res = mask_binary(masked_frame)

     # Apply MOG2 background subtraction
    fg_mask_mog2 = mog2.apply(res)

    # Apply KNN background subtraction
    fg_mask_knn = knn.apply(res)

    # Combine the masks
    fgMask = cv2.bitwise_and(fg_mask_mog2, fg_mask_knn)


    is_detected = False

    results = model(frame, agnostic_nms=True)
    for result in results:
        for r in result.boxes.data.tolist():
            x1, y1, x2, y2, score, class_id = r
            if score > 0.1:
                if class_id == 4: # pedestrian
                    left = int(x1)
                    top = int(y1)
                    right = int(x2)
                    bottom = int(y2)
                if class_id != 4:
                    left_i = int(x1)
                    top_i = int(y1)
                    right_i = int(x2)
                    bottom_i = int(y2)

                    if (left - 30 < left_i < right + 30) and (
                        top - 30 < top_i < bottom + 30
                    ):
                        is_detected = True

                # drawing bboxes
                if class_id == 4:
                    draw_bbox(frame, left, top, right, bottom, class_id, is_detected)
                else:
                    draw_bbox(
                        frame, left_i, top_i, right_i, bottom_i, class_id, is_detected
                    )

    cv2.imshow("frame", frame)
    if cv2.waitKey(1) & 0xFF == ord("q"):
        break
cap.release()
cv2.destroyAllWindows()
```
